name basics
Description : Contient les données personnes (id, nom, date de naissance, date de décés, profession et l'id des films pour chaque personne ).
Colonnes :
nconst : Identifiant unique pour chaque personne
primaryName : le nom de la personne
birthYear : annee de naissance
deathYear : annee de décés
primaryProfession : la profession (une personne peut avoir plusieurs profession)
knownForTitles : id des films



```python
import pandas as pd
import seaborn as sns
import requests
from io import BytesIO
import gzip
import numpy as np
```

```python
url = "https://datasets.imdbws.com/name.basics.tsv.gz"
df_name_basics = pd.read_csv(url, sep="\t", dtype_backend="pyarrow", engine = "pyarrow", on_bad_lines = "skip",usecols =  ['nconst', 'primaryName','primaryProfession', 'knownForTitles'])


```

```python
#création de copier de la dataframe :
df_name_basics_c = df_name_basics.copy()
```

```python
df_name_basics_c
```

```python
#remplacement des \N par les Nan afin de suprimer les lignes avec Nan :

df_name_basics_c['knownForTitles'] = np.where(df_name_basics_c['knownForTitles'] == '\\N', np.nan, df_name_basics_c['knownForTitles'])
```

```python
df_name_basics_c
```

```python
#suppression des lignes avec Nan
df_name_basics_c.dropna(subset=['knownForTitles'], inplace = True)
```

```python
# df name basic copy sans les lignes vides
df_name_basics_c
```

```python
# Filtrer les lignes où la colonne 'primaryProfession' contient "actor", "actress" ou "producer"
filtered_df_name_basics_c = df_name_basics_c[df_name_basics_c['primaryProfession'].str.contains('actor|actress|director', case=False, na=False)]

# Afficher lignes pour vérifier
df_name_basics_c = filtered_df_name_basics_c
print(df_name_basics_c)
```

```python
df_name_basics_c
```

```python
#Vérifier le type de la colonne KnowForTitles
print(df_name_basics_c["knownForTitles"].dtype)
```

```python
#Création d'une copier de la copier pour pouvoir explode la colonne
df_name_basics_c = df_name_basics_c.copy()
#df_name_basics_c["knownForTitles"] = df_name_basics_c["knownForTitles"].fillna('').astype(str)
#print(df_name_basics_c["knownForTitles"].dtype)
```

```python
#Modifier la colonne en str
df_name_basics_c.loc[:, "knownForTitles"] = df_name_basics_c["knownForTitles"].fillna('').astype(str)
print(df_name_basics_c["knownForTitles"].dtype)
```

```python
print(df_name_basics_c["knownForTitles"].apply(type).value_counts())
```

```python
#'knownForTitles' est une chaîne de caractères séparée par des virgules, la convertir en liste
df_name_basics_c["knownForTitles"] = df_name_basics_c["knownForTitles"].apply(lambda x: x.split(',') if isinstance(x, str) else [])

# Exploser la colonne pour que chaque élément de la liste soit une ligne séparée
df_exploded = df_name_basics_c.explode("knownForTitles")

# Vérification du résultat
print(df_exploded.head())
```
