```python
import pandas as pd

akas = '/Users/algaumon/Downloads/Projet_2/VS_code_akas/akas_fr_en_filtered.tsv'
movies = '/Users/algaumon/Documents/Projet_2/VS_code_akas/movies.tsv'
ratings = '/Users/algaumon/Documents/Projet_2/VS_code_akas/title.ratings.tsv'
personne ='/Users/algaumon/Documents/Projet_2/VS_code_akas/personne.csv'

df_akas = pd.read_csv(akas, sep='\t')
df_movies = pd.read_csv(movies, sep='\t')
df_ratings = pd.read_csv(ratings, sep='\t')
df_personne = pd.read_csv(personne,sep=',')
```

```python
df_personne.drop(columns='Unnamed: 0')
```

```python
df_combined = pd.merge(df_movies, df_akas, left_on='tconst', right_on='titleId', how='inner')
```

```python
df_filme = df_combined.drop(columns=['titleType','primaryTitle','titleId','ordering','types','attributes','isOriginalTitle'])
```

```python
df_filmindex= df_filme.reindex(['tconst','originalTitle','title','startYear','isAdult','runtimeMinutes','genres','region','language'],axis = 1)
```

```python
df_sorted = df_filmindex.sort_values(by='region', ascending=False)
df_ok = df_sorted.drop_duplicates(subset='tconst', keep='first')
```

```python
df_ok['startYear'] = pd.to_numeric(df_ok['startYear'], errors='coerce')
```

```python
df_ok['startYear'].fillna(0, inplace=True)
```

```python
df_ok['startYear'] = df_ok['startYear'].astype(int)
```

```python
df_ok=df_ok[df_ok['startYear']!=0]
```

```python
df_ok.sort_values('startYear', ascending=False)
```

```python
condition_start = df_ok['startYear'] > 1959
df_cond = df_ok[condition_start]
```

```python
condition_end = df_cond['startYear'] < 2025
df_top = df_cond[condition_end]
```

```python
df_top.sort_values('startYear',ascending=False, inplace=True)
```

```python
df_top= df_top[df_top['runtimeMinutes']!='\\N']
```

```python
df_top = df_top[df_top['genres'] != r'\N']
```

```python
df_top[df_top['originalTitle'].apply(lambda x: 'Harry Potter' in x)]
```

```python
df_top['runtimeMinutes'] = df_top['runtimeMinutes'].astype(int)
```

```python
df_top = df_top[(df_top['runtimeMinutes'] > 59) & (df_top['runtimeMinutes'] < 220)]
```

```python
df_top=df_top.drop(columns=['region','language'])
```

```python
df_rated = pd.merge(df_top,df_ratings, how='inner',left_on='tconst',right_on='tconst')
```

```python
df_rated[df_rated['originalTitle'].apply(lambda x:'La cité' in x)]
```

```python
df_num = df_rated[(df_rated['numVotes'] > 2000) & (df_rated['averageRating'] > df_rated['averageRating'].mean())]
```

```python
df_rated['averageRating'].mean()
```

```python
df_num=df_num[df_num['genres']!='Documentary']
```

```python
df_num[df_num['originalTitle'].apply(lambda x: 'Superbad' in x)].sort_values('numVotes')
```

```python
df_num.value_counts('isAdult')
```

```python
df_no_docu = df_num[~df_num['genres'].str.contains('documentary', case=False, na=False)]
```

```python
df_no_docu
```

```python
df_no_docu['genres'].value_counts()
```

```python
# Séparer les genres multiples en listes
df_no_docu['genres_split'] = df_no_docu['genres'].str.split(',')

# Aplatir toutes les listes de genres dans une seule série
all_genres = df_no_docu['genres_split'].explode().str.strip()

# Compter les occurrences de chaque genre
genre_counts = all_genres.value_counts()

# Afficher les résultats
print(genre_counts)
```
