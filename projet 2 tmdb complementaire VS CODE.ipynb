```python
# Sélectionne les colonnes souhaitées
colonnes = ['genres', 'id', 'imdb_id', 'original_language', 
            'original_title', 'overview', 'poster_path', 'release_date', 
            'runtime', 'spoken_languages', 'status', 'title', 'video', 
            'vote_average', 'vote_count']

dfnew = df[colonnes]

# Afficher les 20 premières lignes du nouveau DataFrame
#print(dfnew.head(20))

dfnew.columns
```

```python
#filtrer les colonnes , visu valeurs manquantes , doublons
dfnew.value_counts().head(40)
```

```python
# Je vais donc prendre que les films qui dure >= à 85min  ----> df['runtime'].describe()  50% : 85min 
condition = dfnew['runtime'] >= 85
dftime = dfnew[condition]

display(dftime.head(10))
```

```python
# Je selectionne fr et en dans original language
condition1 = dftime["original_language"].isin(["fr","en"])  #autre methode : condition = df["original_language"] == "fr" | df.original_language == "en"
df_en_fr = dftime[condition1]
print(df_en_fr.value_counts())
```

```python
# verification 
df_en_fr['original_language'].unique()
```

```python
#je selectionne les films release date sup à 1960 : release_date
df_en_fr['release_date'].info()
```

```python
#transformation en format datetime :
df_en_fr["release_date"]= pd.to_datetime(df_en_fr["release_date"])
print(df_en_fr.info())
print(df_en_fr['release_date'].head(2))
```

```python
#condition 
condition2 = df_en_fr['release_date'] >= "1960-01-01"
dftime = df_en_fr[condition2]                        #dftime = df_en_fr[df_en_fr['release_date'] >= "1960-01-01"]
display(dftime.head(5))

```

```python
#je selectionne les films deja realisés donc status : released
condition3 = dftime["status"].isin(["Released"])
dfreal = dftime[condition3]
display(dfreal.head())
```

```python
dfreal.shape
```

```python
# je selectionne les films ayant une note sup ou egale à 6/10 vote_average
condition4 = dfreal["vote_average"] >= 6
dfnote = dfreal[condition4]
display(dfnote.head())
```

```python
#verifier si doublons ? 
condition5 = dfnote.duplicated(['imdb_id'],keep=False)
doublons = dfnote[condition5]
print(doublons)

#pas de doublons
```

```python
#verification des valeurs manquantes
dfnote.isnull().sum()
```

```python
#En vue du nombre de films et du nombre de données manquantes , je decide d'enlever les films n'ayant pas de valeur poster_path et pas d'overview
dfnote = dfnote.dropna()
display(dfnote.head())
```

```python
#verification / il ne devrait plus y avoir de valeurs manquantes 
dfnote.isnull().sum()
```

```python
#vu le nombre je vais filtrer en + sur le nombre de votes , avec nbres votes sup à 1000
condition6 = dfnote["vote_count"] >= 1000
dfnote = dfnote[condition6]
display(dfnote.head())
```
