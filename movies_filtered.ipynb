```python
import pandas as pd
import requests
from io import BytesIO
import gzip

# URL du fichier IMDB (par exemple title.basics.tsv.gz)
url = "https://datasets.imdbws.com/title.basics.tsv.gz"

# Télécharger le fichier
response = requests.get(url)
response.raise_for_status()  # Vérifie que le téléchargement s'est bien passé

# Décompresser et charger dans un DataFrame
with gzip.open(BytesIO(response.content), 'rt') as f:
    df = pd.read_csv(f, sep='\t')
```

```python
df
```

```python
df_copy = df.copy()
```

```python
df_movies = df_copy[df_copy['titleType']=='movie']
df_movies.sort_values('startYear', ascending = False)
```

```python
df_col = df_movies.drop(['titleType','endYear'],axis=1)
```

```python
df_col.info()
```

```python
df_star = df_col[df_col['startYear'] != '\\N']
```

```python
df_star[df_star['primaryTitle']=='The Grudge']
```

```python
df_star['startYear'] = pd.to_numeric(df_star['startYear'], errors='coerce')
```

```python
condition_start = df_star['startYear'] > 1936
df_cond = df_star[condition_start]
```

```python
condition_end = df_cond['startYear'] < 2025
df_cond_end = df_cond[condition_end]
```

```python
df_cond_end[df_cond_end['primaryTitle']=='The Grudge']
```
